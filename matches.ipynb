Win/Loss Trends

```python
win_loss=matches.groupby(['season', 'winner']).size().reset_index(name='win_count')
teams=win_loss['winner'].unique()
for team in teams:
    team_data=win_loss[win_loss['winner']==team]
    plt.figure(figsize=(10, 5))
    plt.plot(team_data['season'], team_data['win_count'], marker='o', label=team)
    plt.title(f'Win Trends for {team}')
    plt.xlabel('Season')
    plt.ylabel('Win Count')
    plt.xticks(rotation=45)
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

Toss Decisions Trends for Each Team

```python
for team in teams:
    team_data=matches[(matches['team1']==team) | (matches['team2']==team)]
    plt.figure(figsize=(10, 5))
    sns.countplot(x='season', data=team_data, hue='toss_decision')
    plt.title(f'Toss Decision Trends for {team}')
    plt.xlabel('Season')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    
```

```python
wins_per_venue=matches.groupby(['venue', 'winner']).size().reset_index(name='wins')
matches['team1', 'team2']=matches[['team1', 'team2']].values.tolist()
total_matches_per_venue=(matches.melt(id_vars='venue', value_vars=['team1', 'team2'],var_name='team_type', value_name='team')
                         .groupby(['venue', 'team'])
                         .size()
                         .reset_index(name='total_matches'))
venue_performance=pd.merge(wins_per_venue, total_matches_per_venue, left_on=['venue', 'winner'], right_on=['venue', 'team'])
venue_performance['win_percentage']=(venue_performance['wins']/venue_performance['total_matches'])*100
venue_performance=venue_performance.sort_values(by=['venue','win_percentage'], ascending=[True, False])
venue_performance
```

```python
# Check number of matches classified as Home and Away for each team
classification_check = matches.groupby(['team1', 'home_away']).size().reset_index(name='count')
print(classification_check)
```

```python
matches['home_away'] = matches.apply(
    lambda row: 'Home' if row['venue'] == home_grounds[row['team1']] else 'Away',
    axis=1
)
home_away_performance=matches.groupby(['home_away', 'winner']).size().reset_index(name='wins')
data=home_away_performance.pivot(index='winner', columns='home_away', values='wins').fillna(0)
data['Home'], data['Away'] = data['Away'], data['Home']
teams=data.index
x=np.arange(len(teams))
width=0.35
fig, ax=plt.subplots(figsize=(10, 5))
home=ax.bar(x-width/2, data['Home'], width, label='Home')
away=ax.bar(x+width/2, data['Away'], width, label='Away')
ax.set_xlabel('Teams')
ax.set_ylabel('Wins')
ax.set_title('Home vs Away Wins')
ax.set_xticks(x)
ax.set_xticklabels(teams, rotation=45)
ax.legend()
for bar in [home, away]:
    for b in bar:
        ax.text(b.get_x()+b.get_width()/2, b.get_height(), int(b.get_height()), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
